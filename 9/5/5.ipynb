```python
import numpy as np
```

```python
# SET THIS TO LOAD EXAMPLE DATA INSTEAD
# example = True
example = False

if example:
    fname = 'example.txt'
else:
    fname = 'input.txt'

with open(fname) as f:
    lines = f.readlines()


def load_line(line):
    return [[ int(el) for el in pair.strip().split(',') ] for pair in line.strip().split(' -> ')]

inputs = [load_line(line) for line in lines]
```

```python
# 0 lines
# 1 which coord
# 2 x or y
inputs = np.array(inputs)
# inputs
```

```python
x_max, y_max = inputs[:,:,0].max(), inputs[:,:,1].max()
x_max, y_max
```

```python

line_map = np.full((x_max+1, y_max+1, len(inputs)), 0)
```

```python
def vert_or_hor(line):
    p1, p2 = line
    x1, y1 = p1
    x2, y2 = p2
    return x1 == x2 or y1 == y2

vert_or_hor(inputs[0]), vert_or_hor(inputs[1]), vert_or_hor(inputs[7])
```

```python
def draw_line(line):
    p1, p2 = line.copy()
 
    x_diff, y_diff = p2[0] - p1[0], p2[1] - p1[1]

    out = np.full((x_max+1, y_max+1), 0)
    
    pts = max(abs(x_diff), abs(y_diff))
    x, y = p1
    for ii in range(pts + 1):
#         print(x, y)
#         print(out)
        out[x, y] = 1
        x += int(x_diff/pts)
        y += int(y_diff/pts)
    
    return out

draw_line(inputs[0])
# draw_line(inputs[0])
# draw_line(inputs[7])
```

```python
def draw_line_diag(line):
    p1, p2 = line.copy()
 
    x_diff, y_diff = p2[0] - p1[0], p2[1] - p1[1]

    out = np.full((x_max+1, y_max+1), 0)
    
    pts = max(abs(x_diff), abs(y_diff))
    x, y = p1
    for ii in range(pts + 1):
#         print(x, y)
#         print(out)
        out[x, y] = 1
        x += int(x_diff/pts)
        y += int(y_diff/pts)
    
    return out

draw_line(inputs[-1])
# draw_line(inputs[0])
# draw_line(inputs[7])
```

```python
for ii, line in enumerate(inputs):
#     print(f"{ii, line}")
    if vert_or_hor(line):
#         print(f"{ii, line}")
        line_map[:,:,ii] = draw_line(line)

```
