````python
# fname = 'example.txt'
fname = 'input.txt'
        
with open(fname) as f:
    lines = f.readlines()

alg_lookup = np.array([char == '#' for char in lines[0].strip()])
array = np.array([[char == '#' for char in line.strip()] for line in lines[2:]])
````

````python
alg_lookup
````

````python
x, y = np.where(array)
pairs = list(zip(list(x), list(y)))
pairs = set(pairs)
points = pairs.copy()
````

````python
def convex_hull(points):
    x_min = min((x for x, y in points))
    x_max = max((x for x, y in points))
    y_min = min((y for x, y in points))
    y_max = max((y for x, y in points))
    return (x_min, y_min), (x_max, y_max)

convex_hull(points)
````

````python
def neighbors(ii, jj, points):
#     to_test = [(ii-1, jj-1), (ii, jj-1), (ii+1, jj-1), (ii-1, jj), (ii, jj), (ii+1, jj), (ii-1, jj+1), (ii, jj+1), (ii+1, jj+1)]
    to_test = [(ii + row, jj + col) for row in range(-1, +2) for col in range(-1, +2)]
    bools = [pt in points for pt in to_test]
    return int("".join(['1' if value else '0' for value in bools]), 2)

# neighbors(2, 2, points)
# neighbors(-20, -20, points)
````

````python
%%time

iterations = 2
for _ in range(iterations):
    # find convex hull of points
    (x_min, y_min), (x_max, y_max) = convex_hull(points)
    # expand convex hull by (?) 2 in each direction
    # init next iteration set
    new_points = set()
    # for each point in expanded hull
    for ii in range(x_min - 1, x_max + 1 + 1):
        for jj in range(y_min - 1, y_max + 1 + 1):
#             print((ii, jj))
            # check the 9 points surrounding/including, convert to binary, lookup index
            idx = neighbors(ii, jj, points)
            # next iteration set gets appropriate point included or subtracted
            if alg_lookup[idx]:
                new_points.add((ii, jj))
    points = new_points
````

````python
def find_idx_from_neighbors(ii, jj, points, touched_points=set(), treat_untouched_as_dark=True):
#     to_test = [(ii-1, jj-1), (ii, jj-1), (ii+1, jj-1), (ii-1, jj), (ii, jj), (ii+1, jj), (ii-1, jj+1), (ii, jj+1), (ii+1, jj+1)]
    to_test = [(ii + row, jj + col) for row in range(-1, +2) for col in range(-1, +2)]
    if treat_untouched_as_dark:
        bools = [pt in points for pt in to_test]
    else:
        bools = [((pt in points) or (not pt in touched_points)) for pt in to_test]
    return int("".join(['1' if value else '0' for value in bools]), 2)

# neighbors(2, 2, points)
# neighbors(-20, -20, points)
````
